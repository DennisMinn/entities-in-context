```python
from data_modules import entities
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import json
```

```python
runs = []
with open("data/runs.json", "r") as f:
    for line in f:
        runs.append(json.loads(line))
        
task1 = pd.Series([run["task1"]["accuracy"] for run in runs])
task12 = pd.Series([run["task12"]["accuracy"] for run in runs])
task14 = pd.Series([run["task14"]["accuracy"] for run in runs])
task16 = pd.Series([run["task16"]["accuracy"] for run in runs])
```

```python
df = pd.read_csv("data/entities_metadata.csv")

entities = [run["entity_augmentation"].split("_")[-1] for run in runs]
occurrences = pd.Series([df.entity.entities[entity].occurrences for entity in entities])
token_lengths = pd.Series([df.entity.entities[entity].token_length for entity in entities])
```

```python
token_lengths.describe().round(decimals=4).astype(object)
```

# Task 1

```python
fields = task1.describe().index.tolist()
fields += ["baseline", "median"]
values = task1.describe().tolist()
values += [0.96, task1.median()]

for field, value in zip(fields, values):
    print(f"{field:.<20}: {round(value, 2):.2f}")
```

```python
# create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# first subplot: scatter plot of token lengths vs. accuracy
ax1.scatter(token_lengths, task1)
ax1.axhline(y=0.96, color='red', alpha=0.5)
ax1.set_xlabel('Token lengths')
ax1.set_ylabel('Accuracy')

# second subplot: scatter plot of occurrences vs. accuracy
ax2.scatter(occurrences, task1, color="grey")
ax2.axhline(y=0.96, color='red', alpha=0.5)
ax2.set_xlabel('Occurrences')
ax2.set_ylabel('Accuracy')

# adjust spacing between subplots
plt.subplots_adjust(wspace=0.3)

# show the plot
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,5))

mu, sigma = 0.9, 0.03

# First graph
ax1.hist(task1, alpha=0.7, bins=50, density=True, color='orange')
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
ax1.plot(x, 1/(sigma*np.sqrt(2*np.pi))*np.exp(-(x-mu)**2/(2*sigma**2)), alpha=0.5, color='red')
ax1.set_title('Distribution of Accuracy')
ax1.set_xlabel('Accuracy')
ax1.set_ylabel('Frequency of Accuracy')

# Second graph
ax2.scatter(task1, occurrences, color="grey")
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
ax2.plot(x, 1/(sigma*np.sqrt(2*np.pi))*np.exp(-(x-mu)**2/(2*sigma**2)) * 598.477851, alpha=0.5, color='red')
ax2.set_xlabel('Accuracy')
ax2.set_ylabel('Number of occurrences')

plt.show()
```
